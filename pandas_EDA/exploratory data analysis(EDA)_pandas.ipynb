```python
# Necessary libraries
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

# Loading dataset...

```python
data = pd.read_csv("/kaggle/input/student-score/student_study_scores.csv")
# converting into dataframe
df = pd.DataFrame(data)
print(df.head(3))
```

# Data Overview

```python
# Preview the first few rows of the dataset
print(df.head(4))

print("-------------------------------------------------\n")
# Check the data types of columns
print(df.info())

print("-------------------------------------------------\n")
# Check for missing values
print(df.isnull().sum())
```

# Data Cleaning
* Before starting the analysis, ensure your data is clean.

```python
# Check for duplicates
print(df.duplicated().sum())

# Drop duplicate rows if any
df.drop_duplicates(inplace=True)
```

1. Ensuring Correct Data Types
* Checking the columns have the correct data types (e.g., hours and score should be numeric).

```python
# Convert 'hours' and 'score' to numeric, if necessary
df['Study Hours'] = pd.to_numeric(df['Study Hours'], errors='coerce')
df['Score'] = pd.to_numeric(df['Score'], errors='coerce')
```

# Summary Statistics
* Generating basic statistics of the numerical columns to understand the distribution of data.

```python
# describe data
print(df.describe())
import pandas as pd
```

# Univariate Analysis
Performing univariate analysis (single variables) to understand the distribution of each variable.

```python
# Plot histogram of hours
sns.histplot(df['Study Hours'], kde=True)
plt.title('Distribution of Hours Studied')
plt.xlabel('Study Hours')
plt.ylabel('Frequency')
plt.show()
```

```python
# Plot histogram of scores
sns.histplot(df['Score'], kde=True)
plt.title('Distribution of Scores')
plt.xlabel('Scores')
plt.ylabel('Frequency')
plt.show()
```

# Bivariate Analysis
* Exploring the relationship between the two variables: Score and Study Hours.

```python
# Scatter plot of score vs hours
sns.scatterplot(x='Study Hours', y='Score', data=df)
plt.title('Score vs Hours')
plt.xlabel('Hours Studied')
plt.ylabel('Scores')
plt.show()
```

```python
# Correlation heatmap
sns.heatmap(df.corr(), annot=True, cmap='coolwarm', linewidths=1)
plt.title('Correlation Heatmap')
plt.show()
```

# Outliers Detection
* Checking for any outliers in the data that could affect the analysis

```python
# Boxplot for hours
sns.boxplot(x=df['Study Hours'])
plt.title('Boxplot of Hours')
plt.show()

# Boxplot for scores
sns.boxplot(x=df['Score'])
plt.title('Boxplot of Scores')
plt.show()
```

# Save the Processed Data 
* Finally saving the cleaned and processed data.

```python
# Save the cleaned dataset to a new CSV file
df.to_csv('/kaggle/working/cleaned_student_data.csv', index=False)
```
